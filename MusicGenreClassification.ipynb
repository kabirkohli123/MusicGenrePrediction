```python
Y = df['Class']
```

```python
X = df.drop(['Class'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
Y_predict = lr.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of Linear Regression is:")
print(lr.score(X_test,Y_test))
from sklearn.model_selection import cross_val_score

print("The Cross val score of Linear Regression is:")
print(cross_val_score(lr,X,Y,cv=10))
```

```python
from sklearn.linear_model import LogisticRegression
lrr = LogisticRegression(multi_class='multinomial')
lrr.fit(X_train,Y_train)
```

```python
Y_predict = lrr.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of Logistic Regression is:")
print(lrr.score(X_test,Y_test))
print("The Cross val score of Logistic Regression is:")
print(cross_val_score(lrr,X,Y,cv=10))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train,Y_train)
```

```python
Y_predict = lasso.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of Lasso Regression is:")
print(lasso.score(X_test,Y_test))
print("The Cross val score of Lasso Regression is:")
print(cross_val_score(lasso,X,Y,cv=10))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train,Y_train)
```

```python
Y_predict = ridge.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of Ridge Regression is:")
print(ridge.score(X_test,Y_test))
print("The Cross val score of Ridge Regression is:")
print(cross_val_score(ridge,X,Y,cv=10))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,Y_train)
```

```python
Y_predict = nb.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of Naive Bayes is:")
print(nb.score(X_test,Y_test))
print("The Cross val score of Naive Bayes is:")
print(cross_val_score(nb,X,Y,cv=10))
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train,Y_train)
```

```python
Y_predict = svc.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of Standard Vector Classifier is:")
print(svc.score(X_test,Y_test))
print("The Cross val score of Standard Vector Classifier is:")
print(cross_val_score(svc,X,Y,cv=10))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,Y_train)
```

```python
Y_predict = rfc.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of Random Forest Classifier is:")
print(rfc.score(X_test,Y_test))
print("The Cross val score of Random Forest Classifier is:")
print(cross_val_score(rfc,X,Y,cv=10))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,Y_train)
```

```python
Y_predict = knn.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of KNN is:")
print(knn.score(X_test,Y_test))
print("The Cross val score of KNN is:")
print(cross_val_score(knn,X,Y,cv=10))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,Y_train)
```

```python
Y_predict = xgb.predict(X_test)
print("The first 10 predicted values of Y is:")
print(Y_predict[:10])
print("The first 10 test values of Y is:")
print(np.array(Y_test[:10]))
print("The accuracy of XGB is:")
print(xgb.score(X_test,Y_test))
print("The Cross val score of XGB is:")
print(cross_val_score(xgb,X,Y,cv=10))
```

```python
from sklearn.model_selection import cross_val_score
models = [lr,nb,knn,lasso,ridge,lrr,svc,rfc,xgb]
accuracies =[]

for i in models:
    scores = cross_val_score(i,X,Y,cv=5)
    accuracies.append (np.mean(scores))
```

```python
accuracies
```

```python
from sklearn import metrics

for i in models:
    Y_predict = i.predict(X_test)

    if Y_predict.dtype.kind in 'fc':  # 'f' for floating point, 'c' for complex numbers
        Y_predict = Y_predict.round().astype(int)  # Round and convert to integer

    # Compute the confusion matrix
    matrix = metrics.confusion_matrix(Y_test, Y_predict)
    
    # Plot the confusion matrix
    sns.heatmap(matrix, annot=True, fmt='d', cmap='Blues')
    plt.show()
```
